# Import Python modules and prepare data

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import seaborn as sns

data = pd.read_csv('C:/Users/Wenyangzi Shi/Downloads/Life-Expectancy-and-GDP-Starter/Life-Expectancy-and-GDP-Starter/all_data.csv')
data.head()
```

```python
data.info()
```

```python
print(data.Country.unique())
```

```python
data.describe(include = 'all')
```

```python
# Question 1: Has life expectancy increased over time in the six nations?
sns.pointplot(x = 'Year', y = 'Life expectancy at birth (years)', hue = 'Country', data = data)
plt.xticks(rotation = 30, fontsize = 10)
fig1 = plt.gcf()
fig1.set_size_inches(12, 10)
plt.show()
plt.clf()
```

```python
# Question 2:Has GDP increased over time in the six nations?
sns.pointplot(x = 'Year', y = 'GDP', hue = 'Country', data = data)
plt.xticks(rotation = 30, fontsize = 10)
fig2 = plt.gcf()
fig2.set_size_inches(12, 10)
plt.show()
plt.clf()
```

```python
# Question 3: Is there a correlation between GDP and life expectancy of a country?
## narrow down to get a better sense of the data
data2 = data[(data['GDP'] > 1.735000e+11) & (data['GDP'] < 4.067500e+12)]
## Scatter plot to show the pattern
sns.scatterplot(x = 'GDP', y = 'Life expectancy at birth (years)', data = data2)
plt.show()
plt.clf()
```

```python
# Question 4: What is the average life expectancy in these nations?
## Calculate average life expectancy grouped by country
data_avglife = data.groupby(['Country'])['Life expectancy at birth (years)'].mean()
## Plot the chart
data_avglife.plot(kind='bar', title='Average life expectancy', ylabel='Mean life expectancy',
         xlabel='Country', figsize=(6, 5))
plt.show()
plt.clf()
```

```python
# Question 5: What is the distribution of that life expectancy?
sns.boxplot(x = 'Country', y = 'Life expectancy at birth (years)', data = data)
fig3 = plt.gcf()
fig3.set_size_inches(10, 15)
plt.xticks(rotation = 30, fontsize = 10)
plt.show()
```
